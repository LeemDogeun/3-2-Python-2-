```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

# 데이터셋 불러오기
csv_path = 'spotify.csv'
df = pd.read_csv(csv_path)
print(df.head())

# 데이터 개형 확인
print(df.info())

# 데이터 크기 확인
print(df.shape)

# 상관관계 분석에 적합하지 않은 변수 제거
# artist, song, genre(명목형 데이터)
# explicit(범주형 데이터)
# mode(숫자이지만 범주형)

df.drop(columns=['artist', 'song', 'genre', 'mode', 'explicit'], inplace=True)
print(df.head())

# 중복치 확인
print(df.duplicated().value_counts())

# 중복치 제거
df.drop_duplicates(inplace=True)
print(df.shape)

# Z-score로 이상치 제거
z_scores = np.abs(zscore(df_num, nan_policy='omit'))
df_num = df_num[(z_scores < 3).all(axis=1)]

# 상관계수 계산
corr_matrix = df_num.corr()

# 최대 상관계수의 절댓값이 0.2보다 작은 변수 확인 및 제거
low_corr_vars = [col for col in corr_matrix.columns if corr_matrix[col][corr_matrix[col].index != col].abs().max() < 0.2]
print(f'Low correlation variables: {low_corr_vars}')

df_num_revised = df_num.drop(columns=low_corr_vars)

# 상관계수 히트맵 생성
plt.figure(figsize=(15, 15))
sns.heatmap(df_num_revised.corr(), annot=True, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# # total songs based on genres
fig = px.histogram(df.groupby('genre', as_index=False).count().sort_values(by='song', ascending=False),
                   x='genre', y='song', template='plotly_dark', title='<b>Total songs based on genres</b>')
fig.show()
```

```python
# popular genres based on popularity
fig = px.histogram(df.groupby('genre', as_index=False).sum().sort_values(by='popularity', ascending=False),
                   x='genre', y='popularity', color_discrete_sequence=['green'], template='plotly_dark',
                   title='<b>Popular genres based on popularity</b>')
fig.update_layout()
```

```python
# [물음2] 개선점 제안
## 여기에 코드를 작성해주세요 ##
import pandas as pd
from scipy import stats

# 데이터셋 불러오기
csv_path = 'spotify.csv'
df = pd.read_csv(csv_path)

# Explicity - True, False에 각각 해당하는 데이터 분리
explicit_tracks = df[df['explicit'] == True]['popularity']
non_explicit_tracks = df[df['explicit'] == False]['popularity']

# 평균 계산
mean_explicit = explicit_tracks.mean()
mean_non_explicit = non_explicit_tracks.mean()

# T-test 수행
t_stat, p_value = stats.ttest_ind(explicit_tracks, non_explicit_tracks)

# 검정량 산출
mean_explicit, mean_non_explicit, t_stat, p_value
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 시각화 설정
plt.figure(figsize=(10, 6))

# 박스플롯으로 명시적 및 비명시적 트랙의 인기도 분포 시각화
sns.boxplot(x='explicit', y='popularity', data=df, palette='pastel')
sns.stripplot(x='explicit', y='popularity', data=df, jitter=True, color='gray', alpha=0.5)

# 제목 및 레이블 설정
plt.title('명시적 및 비명시적 트랙의 인기도 비교')
plt.xlabel('명시적 여부')
plt.ylabel('인기도')

# 평균 표시
mean_explicit = df[df['explicit'] == True]['popularity'].mean()
mean_non_explicit = df[df['explicit'] == False]['popularity'].mean()
plt.axhline(mean_explicit, color='blue', linestyle='--', label=f'명시적 평균: {mean_explicit:.2f}')
plt.axhline(mean_non_explicit, color='orange', linestyle='--', label=f'비명시적 평균: {mean_non_explicit:.2f}')

plt.legend()

# 그래프 출력
plt.show()
```

```python
# [물음3] explicity가 popularity에 영향을 주는지 주지 않는지 판단하고 시각화를 통해 이를 정당화하세요.
## 여기에 코드를 작성해주세요 ##
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset and remove columns that are not suitable for correlation analysis
csv_path = 'spotify.csv'
df = pd.read_csv(csv_path)
df = df.drop(columns=['song', 'artist', 'genre', 'mode'])

# Separate the dataset into explicit true and false groups
explicit_true = df[df['explicit'] == True]
explicit_false = df[df['explicit'] == False]

# Calculate the correlation matrix for each group
correlation_matrix_true = explicit_true.corr()
correlation_matrix_false = explicit_false.corr()

# Identify the top 3 variables with the highest correlation to popularity in each group
top3_true = correlation_matrix_true['popularity'].drop('popularity').nlargest(3)
top3_false = correlation_matrix_false['popularity'].drop('popularity').nlargest(3)

# Plotting
plt.figure(figsize=(14, 10))

# Plot for explicit=True group
for i, var in enumerate(top3_true.index):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(x=var, y='popularity', data=explicit_true)
    sns.regplot(x=var, y='popularity', data=explicit_true, scatter=False, color='red', line_kws={'linewidth': 2})
    plt.title(f'Explicit=True: {var} vs Popularity')
    plt.xlabel(var)
    plt.ylabel('Popularity')

# Plot for explicit=False group
for i, var in enumerate(top3_false.index):
    plt.subplot(2, 3, i + 4)
    sns.scatterplot(x=var, y='popularity', data=explicit_false)
    sns.regplot(x=var, y='popularity', data=explicit_false, scatter=False, color='red', line_kws={'linewidth': 2})
    plt.title(f'Explicit=False: {var} vs Popularity')
    plt.xlabel(var)
    plt.ylabel('Popularity')

plt.tight_layout()
plt.show()

```
